```python
import ast
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset

df = pd.read_csv('D:\python practice/Data Analytics/Real Dataset Practice/Student academic performance/student_habits_performance.csv')
```

```python
#filtering the data to gender
df_filtered = df[df['gender'].isin(['Female','Male'])].copy()
```

```python
variables = [
    'study_hours_per_day',
    'sleep_hours',
    'exercise_frequency',
    'exam_score',
    'attendance_percentage'
    
]

titles= ['Study Hours per Day', 'Sleep Hours', 'Exercise Frequency', 'Exam Score','Attendance(%)']
```

```python
fig,ax = plt.subplots(2,3,figsize=(30,20))
ax = ax.flatten()
for i in range(len(variables)):
    sns.histplot(data=df, x=variables[i],hue='gender', kde=True, ax=ax[i], palette='Set2',edgecolor='black')
    ax[i].set_title(titles[i],fontsize=20)
    sns.despine()

if len(ax) > len(variables):
    fig.delaxes(ax[-1])
```
